## FEATURE ENGINEERING

Memperbaikin kolonm biaya pendaftaran

```python
import re

def proses_biaya_mean(biaya_str):
    biaya_str = str(biaya_str).lower()
    
    # Jika gratis/free/strip, anggap 0
    if 'gratis' in biaya_str or 'free' in biaya_str or biaya_str == '-' or biaya_str == '0':
        return 0.0
        
    # Bersihkan karakter 'rp', titik, dan spasi
    biaya_str = biaya_str.replace('rp', '').replace('.', '').replace(' ', '')
    
    # Jika berupa range harga (ada tanda '-')
    if '-' in biaya_str:
        parts = biaya_str.split('-')
        try:
            biaya_min = int(parts[0])
            biaya_max = int(parts[1])
            # Ambil nilai tengah (mean)
            return (biaya_min + biaya_max) / 2
        except ValueError:
            return 0.0
    else:
        # Jika harga tunggal (tidak ada range)
        try:
            harga = int(re.sub(r'\D', '', biaya_str)) # Ambil digit angkanya
            return float(harga)
        except ValueError:
            return 0.0

# Aplikasikan fungsi ke kolom Biaya Registrasi
df['Biaya_Rata_Rata'] = df['Biaya Registrasi'].apply(proses_biaya_mean)

# Hapus kolom aslinya karena sudah digantikan dengan angka mean
df = df.drop(columns=['Biaya Registrasi'])

# Cek hasil perubahannya
print("Preview perubahan biaya menjadi Mean:")
df[['Judul', 'Biaya_Rata_Rata']].head(10)
```

mengubah format tanggal

```python
# CELL 2: Fungsi pemecah menjadi datetime64 murni
import pandas as pd

def parse_tanggal_datetime(tgl_str):
    tgl_str = str(tgl_str).strip()
    
    # Kamus terjemahan ke bahasa Inggris agar Pandas paham format waktunya
    bulan_indo_eng = {
        'Jan': 'Jan', 'Feb': 'Feb', 'Mar': 'Mar', 'Apr': 'Apr', 
        'Mei': 'May', 'Jun': 'Jun', 'Jul': 'Jul', 'Agu': 'Aug', 
        'Sep': 'Sep', 'Okt': 'Oct', 'Nov': 'Nov', 'Des': 'Dec'
    }
    
    # Kalau tidak ada tanda '-', kembalikan nilai Not a Time (NaT)
    if '-' not in tgl_str:
        return pd.Series([pd.NaT, pd.NaT])
        
    try:
        parts = tgl_str.split('-')
        mulai_str = parts[0].strip()
        selesai_str = parts[1].strip()
        
        # Ekstrak tahun
        tahun = selesai_str.split()[-1]
        
        # Skenario 1: Tanggal mulai cuma angka (contoh: "5 - 22 Feb 2026")
        if len(mulai_str.split()) == 1: 
            bulan_selesai = selesai_str.split()[1]
            mulai_str = f"{mulai_str} {bulan_selesai} {tahun}"
            
        # Skenario 2: Tanggal mulai ada bulan, ga ada tahun (contoh: "14 Feb - 10 Apr 2025")
        elif len(mulai_str.split()) == 2:
            mulai_str = f"{mulai_str} {tahun}"
            
        # Terjemahkan ke bahasa Inggris
        for indo, eng in bulan_indo_eng.items():
            # Pakai spasi di sekitar singkatan biar aman replace-nya
            mulai_str = mulai_str.replace(f" {indo} ", f" {eng} ")
            selesai_str = selesai_str.replace(f" {indo} ", f" {eng} ")
            
        # KONVERSI KE DATETIME PANDAS SECARA PAKSA
        tgl_mulai = pd.to_datetime(mulai_str, format='%d %b %Y')
        tgl_selesai = pd.to_datetime(selesai_str, format='%d %b %Y')
        
        return pd.Series([tgl_mulai, tgl_selesai])
        
    except Exception as e:
        return pd.Series([pd.NaT, pd.NaT])

print("✅ Fungsi parse_tanggal_datetime siap digunakan!")
```

```python
# 1. Jalankan fungsi dan pecah jadi 2 kolom baru
df[['Tanggal_Mulai', 'Tanggal_Selesai']] = df['Tanggal Pelaksanaan'].apply(parse_tanggal_datetime)

# 2. Buang kolom tanggal yang lama
df = df.drop(columns=['Tanggal Pelaksanaan'])

# 3. Cek hasil struktur terbaru
print("Struktur Data Terbaru:")
df.info()

print("\nPreview Kolom Tanggal (Format standar YYYY-MM-DD):")
display(df[['Judul', 'Tanggal_Mulai', 'Tanggal_Selesai']].head(10))
```

### KATEGORI/TEMA MAPPING

```python
df2 = pd.read_csv("data_clean.csv")

print("data_clean.csv:")
display(df2.head())
```

```python
# CELL 4: Grouping Domain Kategori (Advanced Level)

# 1. Definisi Mapping Kategori Induk (Super Domain)
# Semua 56 kategori di datamu sudah dimasukkan ke kelompok yang relevan
domain_mapping = {
    'Domain_Akademik': [
        'Esai', 'Olimpiade', 'Karya Tulis Ilmiah', 'Sastra', 'Debat', 'English', 
        'Pidato', 'Akuntansi', 'Media Pembelajaran', 'Beasiswa', 'Artikel', 
        'Hukum', 'Cerdas Cermat', 'Paper', 'Bahasa Asing', 'Try Out', 'Pajak', 
        'Statistika/Data', 'Keuangan', 'Kesehatan'
    ],
    'Domain_Seni_Kreatif': [
        'Desain', 'Videografi/Film', 'Musik', 'Seni', 'Poster', 'Infografis', 
        'Fotografi', 'Dance/Tari', 'Menggambar/Drawing/Ilustrasi', 'Mewarnai', 
        'Fashion Show', 'Story Telling', 'MC/Protocol', 'Voice Over', 
        'News Anchor/Pembawa Berita', 'Podcast'
    ],
    'Domain_Teknologi': [
        'IT', 'Teknik', 'Robot', 'UI/UX'
    ],
    'Domain_Olahraga_E-Sport': [
        'Olahraga', 'E-sport'
    ],
    'Domain_Bisnis_Karir': [
        'Bisnis', 'Seminar', 'Pelatihan', 'Trading', 'Ambassador'
    ],
    'Domain_Umum_Lainnya': [
        'Agama', 'Challenge', '~Lainnya', 'Pramuka', 'Rally Games', 
        'Baris Berbaris', 'Giveaway', 'PMR', 'Permainan', 'Lainnya'
    ]
}

# Bikin dictionary lookup agar mesin cepat mencocokkan katanya
# (Contoh: 'UI/UX' -> 'Domain_Teknologi')
lookup_dict = {}
for domain, kats in domain_mapping.items():
    for k in kats:
        lookup_dict[k] = domain

# 2. Fungsi untuk mengubah deretan kategori mentah menjadi Super Domain
def map_to_domain(teks_kategori):
    teks_kategori = str(teks_kategori).replace(', ', ',')
    kategori_list = teks_kategori.split(',')
    
    domain_terpilih = set() # Pakai set agar tidak ada duplikat domain di 1 lomba
    for kat in kategori_list:
        kat = kat.strip()
        if kat in lookup_dict:
            domain_terpilih.add(lookup_dict[kat])
        elif kat: # Kalau misal ada kategori aneh yang terlewat
            domain_terpilih.add('Domain_Umum_Lainnya') 
            
    # Gabungkan kembali pakai koma (contoh: "Domain_Akademik,Domain_Teknologi")
    return ','.join(list(domain_terpilih))

# 3. Aplikasikan ke kolom Tema/Kategori
df2['Super_Kategori'] = df2['Tema/Kategori'].apply(map_to_domain)

# 4. Lakukan Multi-Label Binarization pada Super Kategori
domain_dummies = df2    ['Super_Kategori'].str.get_dummies(sep=',')

# 5. Gabungkan kolom-kolom baru tersebut ke DataFrame utama
df2 = pd.concat([df2, domain_dummies], axis=1)

# 6. Hapus kolom Tema/Kategori dan Super_Kategori karena sudah jadi angka 0 & 1
df2 = df2.drop(columns=['Tema/Kategori', 'Super_Kategori'])

# Cek hasilnya
print("\n✅ Pemetaan Domain Berhasil!")
print("Bentuk data sekarang:", df2.shape)
display(df2.head())
```

### Memfokuskan kolom jenjang

Jika jenjangnya lebih dari 3 (berarti 4, 5, dst), ubah jadi 'Umum' dan Jika jenjangnya 1, 2, atau 3, ambil elemen yang paling pertama [0]

```python
# Fungsi untuk menyederhanakan kolom Jenjang
def rapihin_jenjang(teks):
    teks = str(teks)
    # Pecah teks berdasarkan koma dan hapus spasi berlebih
    parts = [p.strip() for p in teks.split(',')]
    
    # Jika jenjangnya lebih dari 3 (berarti 4, 5, dst), ubah jadi 'Umum'
    if len(parts) > 3:
        return 'Umum'
    # Jika jenjangnya 1, 2, atau 3, ambil elemen yang paling pertama [0]
    else:
        return parts[0]

# 1. Aplikasikan fungsinya untuk menimpa kolom Jenjang yang lama
df2['Jenjang'] = df2['Jenjang'].apply(rapihin_jenjang)

# 2. Karena isinya sudah sangat rapi dan seragam, kita ubah lagi jadi tipe 'category' 
# (biar hemat memori dan cepat diolah)
df2['Jenjang'] = df2['Jenjang'].astype('category')

# 3. Cek hasil perubahannya
print("Preview Jenjang yang sudah dirapikan:")
display(df2[['Judul', 'Jenjang']].head(10))

# Jangan lupa save ke CSV akhir kalau ini udah tahap paling final ya!
# df2.to_csv('data_clean_final.csv', index=False)
```

### Encoding Tipe(online/offline)

Kalau lombanya "Online" ➡️ Is_Online = 1, Is_Offline = 0

Kalau lombanya "Offline" ➡️ Is_Online = 0, Is_Offline = 1

Kalau lombanya "Online & Offline" ➡️ Is_Online = 1, Is_Offline = 1 (Karena mendukung keduanya!)

```python
# Membuat kolom Is_Online (Kasih nilai 1 jika ada kata 'Online')
df2['Is_Online'] = df2['Tipe (Online/Offline)'].apply(lambda x: 1 if 'Online' in str(x) else 0)

# Membuat kolom Is_Offline (Kasih nilai 1 jika ada kata 'Offline')
df2['Is_Offline'] = df2['Tipe (Online/Offline)'].apply(lambda x: 1 if 'Offline' in str(x) else 0)

# Hapus kolom aslinya karena sudah diwakili oleh 2 kolom angka di atas
df2 = df2.drop(columns=['Tipe (Online/Offline)'])

# Cek hasil perubahannya
print("Preview perubahan Tipe Lomba:")
display(df2[['Judul', 'Is_Online', 'Is_Offline']].head(10))
```
